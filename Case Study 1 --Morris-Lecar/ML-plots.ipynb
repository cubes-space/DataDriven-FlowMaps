```python
errors = sio.loadmat('errors.mat')
```

```python
error_mat = errors.get('mte')
```

```python
error_mat.shape
```

```python
mean_error=np.zeros(error_mat.shape[1], )
std_error=np.zeros(error_mat.shape[1], )
```

```python
for j in range(error_mat.shape[1]):
    mean_error[j] = np.mean(error_mat[:,j])
    std_error[j] = np.sqrt(np.var(error_mat[:,j]))
    
minval = mean_error-std_error
maxval = mean_error+std_error
```

```python
#plt.figure(10,dpi=600)
points = [20,40,60,80,100,120,140,160,180,200,220,240]
ax = plt.figure(dpi=600).gca()
plt.plot(points,mean_error,'o',color=colorlist[2])
ax.vlines(points,minval,maxval, linestyles='solid',color='blue', alpha=0.5)

plt.xlabel(r'$N_s$', fontsize=25)
plt.ylabel(r'$\hat{\epsilon}$',fontsize=25)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
```
