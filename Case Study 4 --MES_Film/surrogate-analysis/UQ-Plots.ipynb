```python
X_opt = inf['X_opt']
X_true = inf['X_true']
```

```python
npars = X_opt.shape[1]
```

```python
mins=np.zeros((npars,))
maxs=np.zeros((npars,))
nominals=np.zeros((npars,))
nominals[0] = 1e-3;nominals[1]=4.5;nominals[2]=0.25;nominals[3]=3.0;nominals[4]=0.1;nominals[5]=2e-5
for i in range(npars):
    mins[i] = nominals[i]-0.1*nominals[i]
    maxs[i] = nominals[i]+0.1*nominals[i]
    print(mins[i],maxs[i])
```

```python
full_samp = 10000000
X_prior_full = np.zeros((full_samp,npars))
for j in range(npars):
    X_prior_full[:,j] = np.random.uniform(mins[j],maxs[j],full_samp)
```

```python
errors = sio.loadmat('average_error_results.mat')
n_simulations = [100,150,200,250,300,350];
```

```python
mean_error = errors.get('mave')
var_error = errors.get('vave')
```

```python
plt.figure(dpi=300)
minus = (mean_error.reshape(6,1) - var_error.reshape(6,1)).reshape(6,)
plus = (mean_error.reshape(6,1) +  var_error.reshape(6,1)).reshape(6,)


plt.plot(n_simulations, mean_error.reshape(6,1), '-o', linewidth=3, markersize=12, color='blue')
plt.fill_between(n_simulations, minus, plus,facecolor='red', alpha=0.2, interpolate=True)

plt.ylabel(r'$\epsilon$', fontsize=20)
plt.xlabel(r'$N_T$',fontsize=18)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.xlim([99,351])
```
